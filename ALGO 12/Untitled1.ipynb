```python
import pandas as pd

df = pd.read_csv(r'/content/sample_data/Negara - Negara.csv')

numeric_cols = df.select_dtypes(include=['number']).columns
rata = df.groupby(['Benua'])[numeric_cols].mean()
std = df.groupby(['Benua'])[numeric_cols].std()

print('BERIKUT DATA FRAMENYA')
print(df)
print('==================BERIKUT DATA MEAN =======================================')
print(rata)
print('==================BERIKUT DATA STANDARD  DEVIATION=======================================')
print(std)
```
